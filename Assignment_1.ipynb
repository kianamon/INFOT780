```python
InitialBoard = np.matrix([[4, 3, 6], [8, 2, 7], [5, 1, 0]])
```

```python
FinalBoard = np.matrix([[8, 1, 2], [7, 0, 3], [6, 5, 4]])
```

Find where the blank tile is located in the begining:

```python
def find_blank(board):
    for i in range(board.shape[0]):
        for j in range(board.shape[1]):
            if board[i, j] == 0:
                index = [i, j]
            else:
                pass
    return index
```

Find all the possible movements we can do:

```python
def find_movement(board):
    i, j = find_blank(board)
    newindices = [[i-1, j], [i+1, j], [i, j+1], [i, j-1]] #up, down, right, left
    validindex = []
    verifyrowindex = [1 if x[0] in range(0, board.shape[0]) else 0 for x in newindices]
    verifycolindex = [1 if x[1] in range(0, board.shape[0]) else 0 for x in newindices]
    for k in range(len(newindices)):
        if verifyrowindex[k] + verifycolindex[k] == 2:
            validindex.append(newindices[k])
    return validindex
```

I am choosing the movement completely random. In other words, we keep moving the tiles randomly to get to the final goal. This may take a while and it is by no means the optimal solution.

```python
def pick_move(movement_list):
    movement = random.choice(movement_list)
    return movement
```

Apply the movement we picked from previous function:

```python
def Apply_Movement(current_board, movement):
    new_board = np.zeros(current_board.shape)
    old_location = find_blank(current_board)
    new_location = movement
    
    for i in range(current_board.shape[0]):
        for j in range(current_board.shape[1]):
            if [i, j] == new_location:
                new_board[i, j] = 0
            elif [i, j] == old_location:
                new_board[i, j] = current_board[new_location[0], new_location[1]]
            else:
                new_board[i, j] = current_board[i, j]
    return np.matrix(new_board)
```

Check to see if we are the final goal or not:

```python
def is_done(current_board, final_board):
    if (current_board == final_board).all():
        return True
    else:
        return False
```

## Main Function:

Solve the puzzle:

```python
def solve_puzzle(initial_board, final_board):
    move_log = []
    if is_done(initial_board, final_board):
        return initial_board
    else:
        current_board = deepcopy(initial_board)
        while is_done(current_board, final_board) == False:
            movetaken = pick_move(find_movement(current_board))
            move_log.append(movetaken)
            current_board = Apply_Movement(current_board, movetaken)
            if is_done(current_board, final_board):
                return (current_board, move_log)
```

```python
finalgoal, movelog = solve_puzzle(InitialBoard, FinalBoard)
```
